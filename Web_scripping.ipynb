```python
#single page info
import requests
from bs4 import BeautifulSoup #to access containt from the website
url='''https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'''
check_status=requests.get(url)
check_status.status_code
print('Successful established a connection status code is',check_status.status_code)
html_data=BeautifulSoup(check_status.content,'html.parser')
#print(html_data)
mobile_name_list=[]
mobile_price_list=[]
mobile_image_list=[]
mobile_feature_list=[]
mobile_names=html_data.find_all("div",class_="KzDlHZ")           #find returns single mobile findall all mobile
for mobile in mobile_names:
    mobile_name_list.append(mobile.text)
    
print(len(mobile_name_list))
mobile_prices=html_data.find_all('div',class_='Nx9bqj _4b5DiR')
for price in mobile_prices:
    mobile_price_list.append(price.text)
print(len(mobile_price_list))
mobile_image=html_data.find_all('div',class_="_4WELSP")
'''for image in mobile_image:
   mobile_image_list.append(image.text)
print(mobile_image_list)'''
mobile_feature=html_data.find_all('div',class_="_6NESgJ")
for feature in mobile_feature:
    mobile_feature_list.append(feature.text)

print(mobile_feature_list)
import pandas as pd
d1={'Mobile Name':mobile_name_list,
    'Mobile Price':mobile_price_list,
    'Mobile Features':mobile_feature_list}
mobile_data=pd.DataFrame(d1)
mobile_data.to_excel('Mobile Info1.xlsx')
```
