```python
#multiple page
import mysql.connector

import requests
from bs4 import BeautifulSoup #to access containt from the website
mobile_name_list=[]
mobile_price_list=[]
mobile_feature_list=[]
page_number=int(input("How many pages information you want"))
for page in range(1,page_number+1):
    url=f'''https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page={page}'''
    check_status=requests.get(url)
    check_status.status_code
    #print('Successful established a connection status code is',check_status.status_code)
    html_data=BeautifulSoup(check_status.content,'html.parser')
    #print(html_data)

    mobile_names=html_data.find_all("div",class_="KzDlHZ")           #find returns single mobile findall all mobile
    for mobile in mobile_names:
        mobile_name_list.append(mobile.text)
        
    #print(len(mobile_name_list))
    mobile_prices=html_data.find_all('div',class_='Nx9bqj _4b5DiR')
    for price in mobile_prices:
        mobile_price_list.append(price.text)
    #print(len(mobile_price_list))
    mobile_image=html_data.find_all('div',class_="_4WELSP")
   
    mobile_feature=html_data.find_all('div',class_="_6NESgJ")
    for feature in mobile_feature:
        mobile_feature_list.append(feature.text)

    #print(mobile_feature_list)
import pandas as pd
d1={'Mobile Name':mobile_name_list,
    'Mobile Price':mobile_price_list,
    'Mobile Features':mobile_feature_list}
#mobile_data=pd.DataFrame(d1)
#mobile_data.to_excel('Mobile Info1.xlsx')
# Connect to MySQL database
x = mysql.connector.connect(
    host='localhost',  # e.g., 'localhost'
    user='root',  # e.g., 'root'
    password='Kamble@123',  # e.g., 'password'
    database='web_scrip'  # e.g., 'mobile_db'
)

y=x.cursor()

# Create table
y.execute('''
    CREATE TABLE IF NOT EXISTS MobileInfo (
        id INT AUTO_INCREMENT PRIMARY KEY,
        mobile_name VARCHAR(255),
        mobile_price VARCHAR(50),
        mobile_features TEXT
    )
''')
print(len(mobile_name_list))
# Insert data into table
for i in range(len(mobile_name_list)):
    y.execute('''
        INSERT INTO MobileInfo (mobile_name, mobile_price, mobile_features)
        VALUES (%s, %s, %s)
    ''', (mobile_name_list[i], mobile_price_list[i], mobile_feature_list[i]))
y.execute('select*from MobileInfo')
data=y.fetchall()
print(data)

# Close the cursor and connection
y.close()
x.close()

print("Data has been inserted into the 'MobileInfo' table.")
```
